```python
import math
import random
import numpy as np
import pandas as pd
from scipy.special import gamma
```

```python
def reccur_weibull_potok(alpha, beta):
    return random.weibullvariate(alpha, beta)
```

```python
MATRIX_COL = 10000
MATRIX_ROW = 10000
```

```python
# Инициализация массива времени
time_list = np.zeros((MATRIX_ROW, MATRIX_COL))

# Генерация mu и sigma сразу для всех строк
alpha_list = np.random.uniform(0.1, 10, size=MATRIX_ROW)
beta_list = np.random.uniform(0.5, 3, size=MATRIX_ROW)

# Заполнение time_list
for i in range(MATRIX_ROW):
    time_values = np.cumsum([reccur_weibull_potok(alpha_list[i], beta_list[i]) for _ in range(MATRIX_COL)])
    time_list[i, :] = time_values

# Создание DataFrame
df = pd.DataFrame(time_list)
df["alpha"] = alpha_list
df["beta"] = beta_list
df["lmbd"] = 1 / (df["alpha"] * gamma(1 + 1 / df["beta"]))
```
