```python
import math
import random
import numpy as np
import pandas as pd
from scipy.special import gamma
```

```python
def reccur_weibull_potok(a, b):
    return random.uniform(a, b)
```

```python
MATRIX_COL = 10000
MATRIX_ROW = 10000
```

```python
# Инициализация массива времени
time_list = np.zeros((MATRIX_ROW, MATRIX_COL))

# Генерация mu и sigma сразу для всех строк
a_list = [] 
b_list = [] 

# Заполнение time_list
for i in range(MATRIX_ROW):
    a = random.uniform(0.01, 1)
    b = random.uniform(a, 10)
    a_list.append(a)
    b_list.append(b)
    time_values = np.cumsum([reccur_weibull_potok(a, b) for _ in range(MATRIX_COL)])
    time_list[i, :] = time_values

# Создание DataFrame
df = pd.DataFrame(time_list)
df["a"] = a_list
df["b"] = b_list
df["lmbd"] = 1 / ((df["a"] + df["b"]) / 2)
```
