```python
import math
import random
import numpy as np
import pandas as pd
```

```python
def reccur_hyperexponential_potok(p, lmbd1, lmbd2):
    # choose_lmbd = random.choices([0, 1], weights=[p, 1 - p])
    # alpha = random.random()
    # time_event = -math.log(alpha) / lmbd_list[choose_lmbd[0]]
    # return time_event
    
    uniform_random = np.random.uniform(0, 1)

    # Определяем, из какого распределения генерировать
    component = uniform_random <= p  # True, если используется λ1, иначе λ2

    # Генерируем случайные числа из соответствующих экспоненциальных распределений
    random_values = np.where(
        component,
        -np.log(np.random.uniform(0, 1)) / lmbd1,  # Если λ1
        -np.log(np.random.uniform(0, 1)) / lmbd2   # Если λ2
    )

    return random_values
```

```python
MATRIX_COL = 10000
MATRIX_ROW = 10000
```

```python
# Инициализация массива времени
time_list = np.zeros((MATRIX_ROW, MATRIX_COL))

# Генерация mu и sigma сразу для всех строк
lmbd1_list = np.random.uniform(0.1, 10, size=MATRIX_ROW)
lmbd2_list = np.random.uniform(0.01, 5, size=MATRIX_ROW)
p_list = [round(random.uniform(0.1, 0.9), 1) for _ in range(MATRIX_ROW)]

# Заполнение time_list
for i in range(MATRIX_ROW):
    time_values = np.cumsum([reccur_hyperexponential_potok(p_list[i], lmbd1_list[i], lmbd2_list[i]) for _ in range(MATRIX_COL)])
    time_list[i, :] = time_values

df = pd.DataFrame(time_list)
# Создание DataFrame
df["lmbd1"] = lmbd1_list
df["lmbd2"] = lmbd2_list
df["p"] = p_list
df["lmbd"] = 1 / ((df["p"] / df["lmbd1"]) + ((1 - df["p"]) / df["lmbd2"]))
```
